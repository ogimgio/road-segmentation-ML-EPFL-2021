## Notebook for working with logistic

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os,sys
from PIL import Image
from sklearn.metrics import accuracy_score
from mega_helpers import *

%load_ext autoreload
%autoreload 2
```

```python
n = min(20, len(files)) # Load maximum 20 images
imgs,gt_imgs = get_images(n)

n = 10 # Only use 10 images for trainin
```

```python
# Extract patches from input images
patch_size = 16 # each patch is 16*16 pixels
#each image has 25*25 possible image patches, so the possible images for all images are 25*25*number of images
img_patches = [img_crop(imgs[i], patch_size, patch_size) for i in range(n)]
gt_patches = [img_crop(gt_imgs[i], patch_size, patch_size) for i in range(n)]
# Linearize list of patches
img_patches = np.asarray([img_patches[i][j] for i in range(len(img_patches)) for j in range(len(img_patches[i]))])
gt_patches =  np.asarray([gt_patches[i][j] for i in range(len(gt_patches)) for j in range(len(gt_patches[i]))])
print(gt_patches.shape)
```

### Preprocessing

```python
X = np.asarray([ extract_features(img_patches[i]) for i in range(len(img_patches))])
Y = np.asarray([value_to_class(np.mean(gt_patches[i])) for i in range(len(gt_patches))])
#each set of values is a pixel, with 3 column since rgb
```

```python
# Print feature statistics

print('Computed ' + str(X.shape[0]) + ' features')
print('Feature dimension = ' + str(X.shape[1]))
print('Number of classes = ' + str(np.max(Y)))  #TODO: fix, length(unique(Y)) 

Y0 = [i for i, j in enumerate(Y) if j == 0]
Y1 = [i for i, j in enumerate(Y) if j == 1]
print('Class 0: ' + str(len(Y0)) + ' samples')
print('Class 1: ' + str(len(Y1)) + ' samples')
```

```python
# train a logistic regression classifier

from sklearn import linear_model

# we create an instance of the classifier and fit the data
logreg = linear_model.LogisticRegression(C=1e4, class_weight="balanced",max_iter=100,solver="liblinear")
logreg.fit(X, Y)


# Run prediction on the img_idx-th image
img_idx = 90
Xi = extract_img_features(image_dir + files[img_idx])
print(Xi.shape)
#the shape is 625 because it is an image so 25 images from 400px to 16x16, but 25*25 poissible one so 625 images
Zi = logreg.predict(Xi)
```

### Accuracy

```python
#calculate accuracy (current)
gt_test_patches = img_crop(load_image(gt_dir + files[img_idx]), patch_size, patch_size)
y_boh = np.asarray([value_to_class(np.mean(gt_test_patches[i])) for i in range(len(gt_test_patches))])
accuracy_score(y_boh, Zi)
```

```python
#predict on set of images (not only on one)
img_idx = 10
Xi = np.empty((0,6))
for i in range(0,10):
    Xi = np.concatenate((Xi,extract_img_features(image_dir + files[i])))
print(Xi.shape)
#the shape is 625 because it is an image so 25 images from 400px to 16x16, but 25*25 poissible one so 625 images
Zi = logreg.predict(Xi)
#calculate accuracy
gt_test_patches = []
for i in range(0,10):
    gt_test_patches = gt_test_patches + img_crop(load_image(gt_dir + files[i]), patch_size, patch_size)
y_boh = np.asarray([value_to_class(np.mean(gt_test_patches[i])) for i in range(len(gt_test_patches))])
accuracy_score(y_boh, Zi)
```
